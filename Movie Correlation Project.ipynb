```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuartion of the plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\User\Desktop\BI Data Analyst\Portfolio Project\movies.csv')
```

```python
#let's look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
#change data type of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Create correct year columns
df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
# Order our Data a little bit to see

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Show everything
pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicate

df['company'].drop_duplicates().sort_values(ascending=False)

# Our option

df.drop_duplicates().head()
```
